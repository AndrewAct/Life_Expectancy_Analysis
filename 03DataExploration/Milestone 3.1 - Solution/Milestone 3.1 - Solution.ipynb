## 2. Data Exploration:  distribution of a single variable

Let's explore the distribution of life expectancy for 2018

```python
df.LifeExp2018.describe()
```

```python
# Which state has the biggest life expectancy?
df[['LifeExp2018','LifeExp2010']][df.LifeExp2018 == max(df.LifeExp2018)]
```

```python
# Which state has the lowest life expectancy?
df[['LifeExp2018','LifeExp2010']][df.LifeExp2018 == min(df.LifeExp2018)]
```

```python
# Plot of the distribution of life expectancy
fig, ax0 = plt.subplots()
sns.distplot(df.LifeExp2018, rug=True, ax=ax0)
ax0.set(title = "Distribution of Life Expectancy (2018)", xlabel = "Life Expectancy in 2018", ylabel = "Density")
ax0.axvline(x=df.LifeExp2018.median(), color='m', label='Median', linestyle='--', linewidth=2)
ax0.axvline(x=df.LifeExp2018.mean(), color='g', label='Mean', linestyle='-', linewidth=2)
ax0.legend()
plt.savefig("./img/LifeExp2018_distplot.png")
plt.show()
```

```python
# Plot the empirical cumulative distribution function (ECDF) of life expectancy
fig, ax0 = plt.subplots()
x = np.sort(df['LifeExp2018'])
y = np.arange(1, len(x)+1)/len(x)
ax0.plot(x, y, '.', linestyle ='none', color='blue')
ax0.set(xlabel='Life expectancy (2018)', ylabel='ECDF', title= 'Empirical Cumulative Distribution\n of Life Expectancy (2018)')
ax0.axvline(x=80, color='g', linestyle='--', linewidth=2)
ax0.margins(0.05)
plt.savefig("./img/LifeExp2018_ECD.png")
plt.show()
```

```python
# ECDF using Statsmodels
from statsmodels.distributions.empirical_distribution import ECDF
fig, ax0 = plt.subplots()
ecdf = ECDF(df['LifeExp2018'])
ax0.plot(ecdf.x,ecdf.y)
ax0.set(xlabel='life expectancy (2018)', ylabel='ECDF', title= 'Empirical Cumulative Distribution\n of Life Expectancy (2018)')
ax0.axvline(x=80, color='g', linestyle='--', linewidth=2)
plt.show()
```

### Answers to the first two questions
1) What percent of the states have a life expectancy greater than 80 years?  
From the ECD plot we see that approximately 80% of the states have a life expectancy of less than 80 years. Consequently, 20% of the states have a life expectancy greater than 80 years.

2) Which state has the highest life expectancy and which state the lowest? 
According to the values of the year 2018:
Hawaii has the largest life expectancy: 82.3   
West Virginia has the lowest life expectancy: 74.8   


## 3. Data Exploration: distribution of a single variable discriminating cases on a categorical variable

Let's see how the life expectancy is distributed across the different regions

```python
# Life Expectancy (2018) statistics across regions
df.groupby('Region')['LifeExp2018'].describe()
```

```python
# kde plots of life expectancy per region
fig, ax0 = plt.subplots()
sns.kdeplot(df.LifeExp2018[df.Region=='North Central'], shade=True,  label='North Central', ax=ax0)
sns.kdeplot(df.LifeExp2018[df.Region=='Northeast'], shade=True, label='Northeast', ax=ax0)
sns.kdeplot(df.LifeExp2018[df.Region=='South'], shade=True, label='South', ax=ax0)
sns.kdeplot(df.LifeExp2018[df.Region=='West'], shade=True,  label='West', ax=ax0)
ax0.set(title = "Distribution of Life Expectancy (2018) per region")
plt.savefig("./img/LifeExp2018_region_kde.png")
plt.show()
```

```python
# box plots of life expectancy per region
g = sns.catplot(x='Region',y='LifeExp2018', kind='box', data=df)
g.fig.suptitle("Life Expectancy per Region")
g.set(ylabel= "life expectancy")
plt.savefig("./img/LifeExp2018_region_box.png")
plt.show()
```

```python
# violinplots of life expectancy per region
g = sns.catplot(x='Region',y='LifeExp2018', kind="violin", data=df)
g.fig.suptitle("Life Expectancy per Region")
g.set(ylabel= "life expectancy")
plt.savefig("./img/LifeExp2018_region_violin.png")
plt.show()
```

```python
# stripplots of life expectancy per region
g = sns.catplot(y='LifeExp2018', x='Region',  kind="strip",data=df)
g.set(ylabel= "life expectancy (years)", xlabel= "region", title="U.S. States")
plt.savefig("./img/LifeExp2018_region_strip.png")
plt.show()
```

### Answer to the third question

3) Is life expectancy equally distributed across the different regions of the U.S.?   
According to the values of the year 2018, life expectancy is not equally distributed across the four regions. The south shows the biggest spread with life expectancy ranging from 74.8 to 80 years.
The south has also the lowest mean at 77.08 years. The other three regions have more similar distributions with means around 79 years.
The west region has the state of Hawaii with the highest life expectancy of 82.3 and the south region the state of West Virginia with the lowest life expectancy of 74.8 years.
